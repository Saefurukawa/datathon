```python
import pandas as pd
import numpy as np
import category_encoders as ce
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

dataset = pd.read_csv("./datathon_train.csv")

def get_time(string):
  return int(string[:2])
dataset['DEP_TIME'] = dataset['DEP_TIME_BLK'].apply(get_time)
```

```python
dataset_clean = dataset.drop(['Id', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME', 'DEP_TIME_BLK', 'DEST'], axis = 1)
dataset_clean.head()
```

```python

# Assuming you have a DataFrame 'X_df' with the data

# Calculate the probability of delayed flights for each unique value in the categorical columns
delayed_probabilities = dataset_clean.groupby('PREVIOUS_AIRPORT')['IS_DELAYED'].mean().to_dict()
delayed_probabilities.update(dataset_clean.groupby('CARRIER_NAME')['IS_DELAYED'].mean().to_dict())
delayed_probabilities.update(dataset_clean).groupby('DEPARTING_AIRPORT')['IS_DELAYED'].mean().to_dict()

# Create a mapping dictionary for each unique value to its corresponding probability
mapping_dict = {
    'PREVIOUS_AIRPORT': delayed_probabilities,
    'CARRIER_NAME': delayed_probabilities,
    'DEPARTING_AIRPORT': delayed_probabilities
}

# Replace the categorical values with their respective probabilities
dataset_clean.replace(mapping_dict, inplace=True)
```
